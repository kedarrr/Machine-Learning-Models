```python
#import libraries
import pandas as pd
```

```python
#load dataset into dataframe
df = pd.read_csv("weather_data.csv")
df
```

```python
#make day as date format from string
df = pd.read_csv("weather_data.csv",parse_dates=["day"])
df
```

```python
#Set the DataFrame index using existing columns
#inplace:- bool, default False : Whether to modify the DataFrame rather than creating a new one
df.set_index("day",inplace=True)
df
```

```python
#Fill 0 value for NaN(Not a Number)
new_df = df.fillna(0)
new_df
```

```python
#Fill appropriate data type for event
new_df = df.fillna({'temperature':0,'windspeed':0,'event':'no event'})
new_df
```

```python
# ffill : fills previous value forward
new_df=df.fillna(method="ffill")
new_df
```

```python
# bfill : fills next value to backword
new_df = df.fillna(method="bfill")
new_df
```

```python
# limit : sets limit on number of changes
new_df = df.fillna(method="ffill",limit=1)
new_df
```

```python
#interpolate : gives avg value to NaN
# method=time : sets values of interpolate according to time
new_df = df.interpolate(method="time")
new_df
```

```python
#drop all rows having NaN values
new_df = df.dropna()
new_df
```

```python
# how=all : drops values if row has all NaN values
new_df = df.dropna(how="all")
new_df
```
